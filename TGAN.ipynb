```python
from tgan.data import load_demo_data
data, continuous_columns = load_demo_data('census')
data.head(3).T
```

```python
from tgan.model import TGANModel

tgan = TGANModel(continuous_columns)
```

```python
tgan.fit(data)
```

now the previous step took toooooooooo long time; so you have to save the resulted model then you can use it any time anywhere

```python
#saving the model
model_path = 'demo/nassar_model'
# Bear in mind that in case the file already exists, 
# TGAN will avoid overwritting it unless the force=True argument is passed:
tgan.save(model_path, force=True)
```

```python
#loading the saved model
new_tgan_nassar = TGANModel.load(model_path)
```

```python
# using the saved model to generate new samples
num_samples = 100

samples = new_tgan_nassar.sample(num_samples)

samples.head(3)

```
